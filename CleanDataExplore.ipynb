## Preprocess Data

### Handle Missing Data

```python
fuel_data.isnull().sum()
```

#### Drop Columns Missing Lots of Data

```python
fuel_data_dropped = fuel_data.drop(['DTI_Status','company_type','weight','tyre_service_time','tyre_service_time_category_A','last_maintenance_date'],axis=1)
```

```python
fuel_data_dropped.head()
```

#### Drop Rows Missing Values in Current Status, Education Level, Truck Driving Experience, Driver Training, and Business Owner

```python
fuel_data_dropped = fuel_data_dropped.dropna(axis=0, subset = ['Current_status', 'Education_level','Truck_driving_experience','driver_trained_before_trip','business_owner'])
```

```python
pd.set_option('display.max_columns', None)
fuel_data_dropped.describe()
```

```python
fuel_data_dropped['Training_Date'].max()
```

#### Impute Mean for Age, Tenure, Vehicle Age, Vehicle Age Days

```python
fuel_data_dropped['Age'].fillna(value=fuel_data_dropped['Age'].mean(),inplace=True)
fuel_data_dropped['Tenure_months'].fillna(value=fuel_data_dropped['Tenure_months'].mean(),inplace=True)
#fuel_data_dropped['vehicle_age'].fillna(value=fuel_data_dropped['vehicle_age'].mean(),inplace=True)
fuel_data_dropped['vehicle_age_days'].fillna(value=fuel_data_dropped['vehicle_age_days'].mean(),inplace=True)
fuel_data_dropped['days_since_last_maintenance'].fillna(value=1000,inplace=True)
```

```python
fuel_data_dropped.isnull().sum()
```

```python
fuel_data_dropped['model'].describe()
```

### Add Independent Vars

#### Make Training Date Datetime Format

```python
fuel_data_subset.Training_Date = fuel_data_subset.Training_Date.apply(lambda x: x.date())
```

#### Fill Missing Dates

```python
fuel_data_subset['Training_Date'].fillna(value=date(2010, 9, 12),inplace=True)
```

#### Reset Index

```python
fuel_data_plus = fuel_data_subset.reset_index()
```

#### Loop Through and Create 'days_since_last_training'

```python
days_since_training = []
```

```python
for i, row in fuel_data_plus.iterrows():
    delta = date(2018, 6, 19) -  row['Training_Date']
    days_since_training.append(delta.days)
```

```python
fuel_data_plus['days_since_last_training'] = days_since_training
```

```python
fuel_data_plus.describe()
```

#### Add Days Since Training Derivatives

```python
training_past_week = []
training_past_month = []
training_past_year = []
training_past_2_year = []
training_greater_2_year = []
```

```python
for value in fuel_data_plus['days_since_last_training']:
    if value <= 7:
        training_past_week.append('Yes')
    else:
        training_past_week.append('No')
```

```python
for value in fuel_data_plus['days_since_last_training']:
    if (7<value <= 31):
        training_past_month.append('Yes')
    else:
        training_past_month.append('No')
```

```python
for value in fuel_data_plus['days_since_last_training']:
    if (31<value <= 365):
        training_past_year.append('Yes')
    else:
        training_past_year.append('No')
```

```python
for value in fuel_data_plus['days_since_last_training']:
    if (365<value <= 730):
        training_past_2_year.append('Yes')
    else:
        training_past_2_year.append('No')
```

```python
for value in fuel_data_plus['days_since_last_training']:
    if (730<value <= 1350):
        training_greater_2_year.append('Yes')
    else:
        training_greater_2_year.append('No')
```

```python
fuel_data_plus['training_past_week'] = training_past_week 
fuel_data_plus['training_past_month'] = training_past_month
fuel_data_plus['training_past_year'] = training_past_year
fuel_data_plus['training_past_2_year'] = training_past_2_year
fuel_data_plus['training_greater_2_year'] = training_greater_2_year
```

```python
fuel_data_plus.head()
```

#### Add Days Since Last Maintenance Derivatives

```python
maintenance_past_week = []
maintenance_past_month = []
maintenance_past_6_month = []
maintenance_past_year = []
maintenance_greater_year = []
```

```python
for value in fuel_data_plus['days_since_last_maintenance']:
    if value <= 7:
        maintenance_past_week.append('Yes')
    else:
        maintenance_past_week.append('No')
```

```python
for value in fuel_data_plus['days_since_last_maintenance']:
    if (7<value <= 31):
        maintenance_past_month.append('Yes')
    else:
        maintenance_past_month.append('No')
```

```python
for value in fuel_data_plus['days_since_last_maintenance']:
    if (31<value <= 183):
        maintenance_past_6_month.append('Yes')
    else:
        maintenance_past_6_month.append('No')
```

```python
for value in fuel_data_plus['days_since_last_maintenance']:
    if (183<value <= 365):
        maintenance_past_year.append('Yes')
    else:
        maintenance_past_year.append('No')
```

```python
for value in fuel_data_plus['days_since_last_maintenance']:
    if (365<value <= 542):
        maintenance_greater_year.append('Yes')
    else:
        maintenance_greater_year.append('No')
```

```python
fuel_data_plus['maintenance_past_week'] = maintenance_past_week
fuel_data_plus['maintenance_past_month'] = maintenance_past_month
fuel_data_plus['maintenance_past_6_month'] = maintenance_past_6_month
fuel_data_plus['maintenance_past_year'] = maintenance_past_year
fuel_data_plus['maintenance_greater_year'] = maintenance_greater_year
```

```python
from sklearn.model_selection import cross_validate, KFold
from collections import defaultdict
```

### KNN Regressor

```python
kf=KFold(n_splits=10, shuffle=True, random_state=11)

knn_r2=defaultdict(list)
for metric in ['uniform','distance']:
    for i in range(1,40):
        #kf=KFold(n_splits=5, shuffle=True, random_state=i+1)
        knn=KNeighborsRegressor(n_neighbors=i,weights=metric)
        cv_knn=cross_validate(knn,X_sc,y,cv=kf,scoring={'r2'},return_estimator=True)
        knn_r2[metric].append(cv_knn['test_r2'].mean())
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,8))

plt.plot(range(1,10),knn_r2['uniform'], label='Weight: Uniform')
plt.plot(range(1,10), knn_r2['distance'],label="Weight: Inv Distance")
plt.xlabel("K",fontsize=14)
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0.00000001,0.2,0.04))
plt.ylabel("Mean R2 Score",fontsize=14)
plt.legend()
plt.show()
```
